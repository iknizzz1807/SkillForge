```python
import pandas as pd
import numpy as np
import string
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.stem import PorterStemmer, WordNetLemmatizer
import re

stop_words = set(stopwords.words('english'))
```

```python
job_data = pd.read_csv('job_title_des.csv')
job_data = job_data.dropna()
job_data = job_data.drop_duplicates()
job_data = job_data.reset_index(drop=True)
job_data = pd.DataFrame(job_data['Job Description'])
job_data['0'] = range(0, len(job_data))
job_data = job_data.rename(columns={'Job Description': '1'})
print(job_data)
```

```python
resume_data = pd.read_csv('resumes.csv')
resume_data = resume_data.dropna()
resume_data = resume_data.drop_duplicates()
resume_data = resume_data.reset_index(drop=True)
resume_data = pd.DataFrame(resume_data['Resume'])
resume_data['0'] = range(0, len(resume_data))
resume_data = resume_data.rename(columns={'Resume': '1'})
resume_data.head()
```

```python
def nlp_preprocess(text, remove_stopwords=True, stemming=False, lemmatization=True):
    """Complete NLP preprocessing pipeline"""
    # Lowercase and remove punctuation
    text = text.lower()
    text = text.translate(str.maketrans('', '', string.punctuation))
    stemmer = PorterStemmer()
    lemmatizer = WordNetLemmatizer()
    # Tokenize
    tokens = word_tokenize(text)
    
    # Remove stopwords if requested
    if remove_stopwords:
        tokens = [word for word in tokens if word not in stop_words]
    
    # Apply stemming if requested
    if stemming:
        tokens = [stemmer.stem(word) for word in tokens]
    
    # Apply lemmatization if requested
    if lemmatization:
        tokens = [lemmatizer.lemmatize(word, pos='v') for word in tokens]
    
    return " ".join(tokens)
```

```python
job_data['1'] = job_data['1'].apply(nlp_preprocess)
resume_data['1'] = resume_data['1'].apply(nlp_preprocess)
```

```python
job_data.to_csv('preprocessed_job_data.csv', index=False)
resume_data.to_csv('preprocessed_resume_data.csv', index=False)
```
